# Tratando dados aninhados e em listas

```python
import pandas as pd
```

```python
dados = pd.read_json('https://raw.githubusercontent.com/brendampa/Pandas_Transf_Manipulacao_Dados.ipynb/refs/heads/main/dados_hospedagem.json')
```

```python
dados.head()
```

```python
dados = pd.json_normalize(dados['info_moveis'])
```

```python
dados
```

```python
#Retirando o agrupamento de listas utilizando o método Explode

# 1. Coletar os nomes das colunas e inserir em uma lista para aplicar no método
colunas = list(dados.columns)
colunas
```

```python
#Ajustando o dataframe dados para que retire o agrupamento apenas das colunas a partir do índice 3:
dados = dados.explode(colunas[3:])
dados
```

```python
#Ajustando os índices com ordenação correta: Drop é para excluir os índices antigos e o inplace é para criar novos
dados.reset_index(inplace=True, drop=True)
dados
```

#Tratando dados numéricos

```python
#Verificar os tipos de dados do DataFrame
# O object é usado quando vários tipos de dados é utilizado ou quando tem um tipo específico mas na leitura na não foi identificado ou quando todos são strings.
dados.info()
```

```python
#Conversão dos tipos de dados
#Método asType - Necessário importar a biblioteca Numpy para utilizar os tipos int64 e float

import numpy as np
```

```python
#Astype converte toda a coluna em um tipo de dado

dados['max_hospedes'] = dados['max_hospedes'].astype(np.int64)
```

```python
dados.info()
```

```python
#Fazendo o mesmo para as demais colunas:
col_numericas = ['quantidade_banheiros','quantidade_quartos','quantidade_camas']
```

```python
dados[col_numericas] = dados[col_numericas].astype(np.int64)
```

```python
dados.info()
```

```python
#Utilizando o float64

dados['avaliacao_geral'] = dados['avaliacao_geral'].astype(np.float64)
```

```python
dados.info()
```

```python
#Conversão com moedas - String $ + números:
# 1. Transformando os dados com apply
#2.Atribuir à coluna Preco
dados['preco'] = dados['preco'].apply(lambda x: x.replace('$','').replace(',','').strip())
```

```python
dados['preco'] = dados['preco'].astype(np.float64)
```

```python
dados.info()
```

```python
#Método para transformar mais de uma coluna ao mesmo tempo: applymap
dados[['taxa_deposito','taxa_limpeza']] = dados[['taxa_deposito','taxa_limpeza']].applymap(lambda x: x.replace('$','').replace(',','').strip())
```

```python
dados[['taxa_deposito','taxa_limpeza']] = dados[['taxa_deposito','taxa_limpeza']].astype(np.float64)
```

```python
dados.info()
```

```python
dados
```

# Tratando dados textuais

```python
dados
```

```python
# Tokenização: segmentar palavras como em uma lista. Exemplo: ['divide','um','texto','em','unidades','menores']
#Foco na coluna descricao_local
dados['descricao_local']
```

```python
#1.Padronizar o texto em letras minúsculas com o método str e seus métodos:
dados['descricao_local'] = dados['descricao_local'].str.lower()
dados['descricao_local']
```

```python
#Remoção de carcateres que não agregam informações
#Regex (ou expressão regular) é uma sequência de caracteres que define um padrão de busca em um texto.

dados['descricao_local'][3169]
```

```python
#Em textos em inglês, é necessário manter o - em palavras compostas  e '
dados['descricao_local'] = dados['descricao_local'].str.replace('[^a-zA-Z0-9\-\']', ' ',regex=True)
```

```python
#Remoção de hífens que não esteja em palavras compostas
dados['descricao_local'] = dados['descricao_local'].str.replace('(?<!\w)-(?!\w)', ' ',regex=True)
dados['descricao_local']
```

```python
#Realizando a TOKENIZAÇÃO
#1. Separando as palavras
dados['descricao_local'] = dados['descricao_local'].str.split()
```

```python
dados.head()
```

```python
#Tratando a coluna Comodidades
#Remoção de caracteres

dados['comodidades'] = dados['comodidades'].str.replace('\{|}|\"','',regex=True)
dados['comodidades']
```

```python
#Criação do token com split
dados['comodidades'] = dados['comodidades'].str.split(',')
```

```python
dados.head()
```

```python
#Coluna descricao_vizinhanca
# Transformamos o texto em letras minúsculas
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.lower()
# Substituimos os caractere especiais
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.replace('[^a-zA-Z0-9\-\']', ' ', regex=True)
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.replace('(?<!\w)-(?!\w)', '', regex=True)
# Transformamos o texto em lista, formando o token
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.split()
dados.head()
```
