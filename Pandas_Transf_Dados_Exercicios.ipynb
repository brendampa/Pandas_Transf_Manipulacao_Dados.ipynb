#DESAFIO 1

```python
import pandas as pd
```

```python
dados_loc = pd.read_json('https://raw.githubusercontent.com/brendampa/Pandas_Transf_Manipulacao_Dados.ipynb/refs/heads/main/dados_locacao_imoveis.json')
```

```python
dados_loc.head()
```

```python
dados_loc = pd.json_normalize(dados_loc['dados_locacao'])
```

```python
dados_loc.head()
```

```python
colunas = list(dados_loc.columns)
colunas
```

```python
dados_loc = dados_loc.explode(colunas[1:])
dados_loc
```

```python
dados_loc.reset_index(inplace=True,drop=True)
dados_loc
```

```python
dados_loc.info()
```

```python
dados_loc['valor_aluguel'] = dados_loc['valor_aluguel'].apply(lambda x: x.replace('$','').replace(',','').replace('reais','').strip())
dados_loc
```

```python
import numpy as np
```

```python
dados_loc['valor_aluguel'] = dados_loc['valor_aluguel'].astype(np.float64)
```

```python
dados_loc.info()
```

```python
# Opção de substituição - necessário verificar o resultado da substituição
# Foi necessário adicionar a barra '\' para ser considerados os parênteses como caracteres literais
# Realizar a substituição dos dados na coluna textual
dados_loc['apartamento'] = dados_loc['apartamento'].str.replace(' \(blocoAP\)', '', regex=True)
# Visualizar o resultado final
dados_loc
```

```python
dados_loc.info()
```

```python
dados_loc['datas_combinadas_pagamento'] = pd.to_datetime(dados_loc['datas_combinadas_pagamento'], format='%d/%m/%Y')
dados_loc['datas_de_pagamento'] = pd.to_datetime(dados_loc['datas_de_pagamento'], format='%d/%m/%Y')
dados_loc.info()
```

```python
dados_loc['Dias_atraso'] = dados_loc['datas_de_pagamento'] - dados_loc['datas_combinadas_pagamento']
dados_loc
```

```python
media_atrasos = dados_loc.groupby(['apartamento'])['Dias_atraso'].mean()
media_atrasos
```

# DESAFIO 2

```python
dados_vendas = pd.read_json('https://raw.githubusercontent.com/brendampa/Pandas_Transf_Manipulacao_Dados.ipynb/refs/heads/main/dados_vendas_clientes.json')
```

```python
dados_vendas.head()
```

```python
dados_vendas = pd.json_normalize(dados_vendas['dados_vendas'])
```

```python
dados_vendas.head()
```

```python
colunas2 = list(dados_vendas.columns)
colunas2
```

```python
dados_vendas = dados_vendas.explode(colunas2[1:])
dados_vendas
```

```python
dados_vendas['Valor da compra'] = dados_vendas['Valor da compra'].apply(lambda x: x.replace('R','').replace(',','.').strip())
dados_vendas
```

```python
dados_vendas['Valor da compra'] = dados_vendas['Valor da compra'].astype(np.float64)
```

```python
dados_vendas.info()
```

```python
# Transformar os textos de Cliente para texto em minúsculo
dados_vendas['Cliente'] = dados_vendas['Cliente'].str.lower()
# Opção de substituição - necessário verificar o resultado da substituição
# O regex não seleciona todas as letras de a-z e espaços em branco ' '
# Tudo que satisfaz o regex é apagado
# Realizar a substituição dos dados na coluna textual
dados_vendas['Cliente'] = dados_vendas['Cliente'].str.replace('[^a-z ]', '', regex=True).str.strip()

# Visualizar o resultado final
dados_vendas.head()
```

```python
dados_vendas.info()
```

```python
dados_vendas['Data de venda'] = pd.to_datetime(dados_vendas['Data de venda'], format='%d/%m/%Y')
```

```python
dados_vendas.info()
```

```python
dados_vendas['Valor da compra'] = dados_vendas['Valor da compra'].apply(lambda x: x.replace('$','').replace(',','').strip())
dados_vendas
```

```python
dados_vendas['Valor da compra'] = dados_vendas['Valor da compra'].astype(np.float64)
dados_vendas.info()
```

```python
vendas = dados_vendas.groupby(['Cliente'])['Valor da compra'].sum()
vendas
```
